# Problem Statement : whether the relevant customer has left the company

## Data Fetching

```python
import pandas as pd
```

```python
df=pd.read_csv('D:/e_commerce.csv')
df
```

# EDA

```python
df.isna().sum()
```

# Define x and y

```python
x=df.drop(['churn'],axis=1)
y=df['churn']
```

# Import Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

# Preprocessing

```python
cat=[]
con=[]

for i in x.columns:
    if x[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),('scaler',StandardScaler())])
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encode',OrdinalEncoder())])
```

# combine cat and con data

```python
pre=ColumnTransformer([('cat_pipe',cat_pipe,cat),('num_pipe',num_pipe,con)])
```

```python
x1=pre.fit_transform(x)
x1
```

```python
x1=pd.DataFrame(pre.fit_transform(x),columns=pre.get_feature_names_out())
x1
```

# Split Data into Training and Testing Set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=23)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
la=LogisticRegression()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
kn=KNeighborsClassifier()
sv=SVC()
```

```python
list1=[la,dt,rf,ab,kn,sv]
```

```python
for i in list1:
    i.fit(x_train,y_train)
    from sklearn.metrics import f1_score,accuracy_score
    y_pred=i.predict(x_test)
    f1=round(f1_score(y_pred,y_test),2)
    acc=round(accuracy_score(y_pred,y_test),2)
    print('*'*50)
    print(i)
    print('Accuracy:',acc)
    print('F1_score:',f1)
    print('Diffrenece:',round((acc-f1),2))
```
